```python
stats = {}
for i in range(0, 97, 3):
    state = sheetX.iloc[i, 0]
    perc = sheetX.iloc[i+2,2:]/sheetX.iloc[i+1,2:]
    av = perc.sum()/10
    av = 1 - av
    stats[state] = [perc, av]
#     print('{0}\t\t\t{1}'.format(state, av))
#     print(i)
```

```python
df2 = pd.DataFrame(stats)
```
